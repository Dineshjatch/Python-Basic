Python Basic

1. What is Python, and why is it popular?
  - Python is a high level programming language. It is widely used in web development, data science and more. Its popularity comes from its readable syntax, extensive libraries, cross-platform compatibility, and strong community support.
2. What is an interpreter in Python?
  - A Python interpreter is a program that executes Python code line by line at runtime, without needing compilation. It translates Python code into machine-readable instructions and allows interactive execution.
3. What are pre-defined keywords in Python?
  - Pre-defined keywords in Python are reserved words that have special meanings and cannot be used as variable names. Examples are if, else, while, for, def, class, return, import, True, False, etc. Python has 33+ keywords.
4. Can keywords be used as variable names?
  - No, keywords cannot be used as variable names in Python because they have predefined meanings and functions in the language.
5. What is mutability in Python ?
  - Mutability in Python refers to whether an object can be changed after creation.
    - Mutable objects (can be modified): lists, dictionaries, sets
    - Immutable objects (cannot be modified): tuples, strings, numbers
6. Why are lists mutable, but tuples are immutable?
  - Lists are mutable because their elements can be changed, added, or removed after creation.
  Tuples are immutable because they are designed to be fixed and unchangeable, ensuring data integrity and better performance.
7. What is the difference between “==” and “is” operators in Python?
  - - == (Equality Operator): Checks if values of two objects are equal.
    - is (Identity Operator): Checks if two objects refer to the same memory location.
8. What are logical operators in Python?
  - Logical operators in Python:

    - and - True if both conditions are true.
    - or - True if at least one condition is true.
    - not - Reverses the boolean value

9. What is type casting in Python?
  - Type casting in Python is converting one data type into another.

     Types:

    - Implicit Casting - Done automatically by Python (e.g., int to float).
    - Explicit Casting - Done manually using int(), float(), str(), etc.

10. What is the difference between implicit and explicit type casting?
  - - Implicit Casting - Done automatically by Python (e.g., int to float).
     - Explicit Casting - Done manually using int(), float(), str(), etc.

11. What is the purpose of conditional statements in Python?
  - Conditional statements in Python control the program flow based on conditions using if, elif, and else to make decisions.

12. How does the elif statement work?
  - The elif statement in Python checks multiple conditions. If the if condition is false, elif is evaluated.

13. What is the difference between for and while loops?
  - - for loop - Used for iterating over a sequence (e.g., list, string, range).
      - while loop - Runs as long as a condition is True.

14. Describe a scenario where a while loop is more suitable than a for loop.
  - A while loop is better when the number of iterations is unknown and depends on a condition.

     Example: Keep asking for a correct password.
